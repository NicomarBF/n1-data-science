## Dia da semana com placares mais altos

Crie um gráfico que responda em qual dia da semana os placares foram mais altos. O conceito de "placar mais alto" é a soma dos gols feitos no jogo.

Como a quantidade de jogos não é a mesma em todos os dias da semana (finais de semana tendem a ter mais jogos que em dias da semana - o que resultaria em placares mais altos nestes dias), você deverá encontrar uma forma de equalizar esta diferença na quantidade de jogos.

A resposta deverá ser um gráfico que demonstre claramente o comparativo dos dias.

````python
matches_file_path = '/content/matches.csv'
matches_df = pd.read_csv(matches_file_path)

day_mapping = {
    'Mon': 'Monday',
    'Tue': 'Tuesday',
    'Wed': 'Wednesday',
    'Thu': 'Thursday',
    'Fri': 'Friday',
    'Sat': 'Saturday',
    'Sun': 'Sunday'
}
matches_df['day'] = matches_df['day'].map(day_mapping)

matches_df['total_goals'] = matches_df['gf'] + matches_df['ga']

goals_per_day = matches_df.groupby('day').agg(
    total_goals_sum=('total_goals', 'sum'),
    total_games=('day', 'count')
).reset_index()

goals_per_day['avg_goals_per_game'] = goals_per_day['total_goals_sum'] / goals_per_day['total_games']

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
goals_per_day['day'] = pd.Categorical(goals_per_day['day'], categories=day_order, ordered=True)
goals_per_day = goals_per_day.sort_values('day')

fig = px.bar(goals_per_day, x='day', y='avg_goals_per_game', title='Média de Gols por Jogo em Cada Dia da Semana',
             labels={'avg_goals_per_game': 'Média de Gols por Jogo', 'day': 'Dia da Semana'},
             color='avg_goals_per_game', color_continuous_scale='Blues')

fig.update_traces(marker_line_color='black', marker_line_width=1.5)

fig.show()
````

## Formação Tática

Qual a formação tática mais comum para cada equipe de acordo com o resultado obtido (Ex: para vitórias, a formação mais comum é X, para empates Y e para Z em derrotas).

**Importante** responda de forma que seja possível fazer comparativos entre equipes.

Responda com um gráfico.

````python
matches_file_path = '/content/matches.csv'
data = pd.read_csv(matches_file_path)

result_mapping = {'W': 'Vitória', 'D': 'Empate', 'L': 'Derrota'}
data['result_desc'] = data['result'].map(result_mapping)

formation_counts = data.groupby(['team', 'result_desc', 'formation']).size().reset_index(name='count')

most_common_formations = formation_counts.loc[formation_counts.groupby(['team', 'result_desc'])['count'].idxmax()]

# Criar o gráfico
fig = px.bar(
    most_common_formations,
    x='team',
    y='count',
    color='formation',
    facet_col='result_desc',
    title='Formação Tática Mais Comum por Resultado para Cada Equipe',
    labels={'count': 'Número de Ocorrências', 'formation': 'Formação'},
    height=600
)

fig.show()
````

## Posse de bola

Qual a média de posse de bola da equipe mandante quando ela perdeu o jogo?

Responda em forma de gráfico para poder comparar entre todas as equipes.

````python
# Carregar os dados
matches_file_path = '/content/matches.csv'
matches_df = pd.read_csv(matches_file_path)

home_losses = matches_df[(matches_df['gf'] < matches_df['ga']) & (matches_df['venue'] == 'Home')]

average_possession = home_losses.groupby('team')['poss'].mean().reset_index()

average_possession_sorted = average_possession.sort_values(by='poss', ascending=False)

fig = px.bar(
    average_possession_sorted,
    x='team',
    y='poss',
    color='poss',
    title='Média de Posse de Bola da Equipe Mandante em Derrotas por Equipe',
    labels={'poss': 'Média de Posse de Bola (%)', 'team': 'Equipe'},
    color_continuous_scale='reds',
    height=600
)

fig.update_traces(marker_line_color='black', marker_line_width=1.5)

fig.update_layout(
    xaxis_title='Equipe',
    yaxis_title='Média de Posse de Bola (%)',
    xaxis_tickangle=-70,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    title_font_size=16
)

fig.show()
````

## Classificação

Qual a classificação final do campeonato?

Responda em forma de tabela, ordenada do primeiro colocado ao último.
Você deverá apresentar o brasão do time (pode ser obtido online [aqui](https://logodetimes.com/premier-league/), mas automatize a busca, não salve o logo localmente), o nome, a quantidade de pontos e a posição final no campeonato.

**OBS** Vitória vale 3 pontos, empates 1 ponto e derrota 0 pontos.

````python
# Responda a etapa aqui
matches_file_path = '/content/matches.csv'
matches_df = pd.read_csv(matches_file_path)

def calculate_points(row):
    if row['gf'] > row['ga']:
        return 3  # vitória
    elif row['gf'] == row['ga']:
        return 1  # empate
    else:
        return 0  # derrota

matches_df['points'] = matches_df.apply(calculate_points, axis=1)

standings = matches_df.groupby('team')['points'].sum().reset_index()

standings = standings.sort_values(by='points', ascending=False).reset_index(drop=True)

standings['position'] = range(1, len(standings) + 1)

print(standings)
````

Crie uma visualização que apresente o público total de cada equipe, e também um detalhamento por jogos.

````python
matches_file_path = '/content/matches.csv'
df = pd.read_csv(matches_file_path)

total_attendance_per_team = df.groupby('team')['attendance'].sum().reset_index()

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id='total-attendance-graph'),
    dcc.Graph(id='game-detail-graph')
])

@app.callback(
    Output('total-attendance-graph', 'figure'),
    [Input('game-detail-graph', 'clickData')])
def update_total_attendance_graph(clickData):
    fig = px.bar(total_attendance_per_team, x='team', y='attendance', title='Público Total por Equipe')
    return fig

@app.callback(
    Output('game-detail-graph', 'figure'),
    [Input('total-attendance-graph', 'clickData')])
def update_game_detail_graph(clickData):
    team = clickData['points'][0]['x'] if clickData else 'Manchester City'
    df_filtered = df[df['team'] == team]
    fig = px.bar(df_filtered, x='date', y='attendance', title=f'Detalhamento de Público por Jogo - {team}')
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
````
